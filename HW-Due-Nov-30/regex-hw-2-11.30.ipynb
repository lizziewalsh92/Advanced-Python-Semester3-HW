```python
## import libraries
import pandas as pd
import regex as re
```

```python
with open("confession-judgments-copy.txt", "r") as text_obj:
    text = text_obj.read()
```

```python
text
```

```python
## write the regex commands

case_num_reg = re.compile(r"PM INDEX NO.\s(\d+)", re.M)
date_received_reg = re.compile(r"NYSCEF:\s(\d{1,2}/\d{1,2}/\d+)", re.M)
amt_reg = re.compile(r"\(\$([\d,]+\.\d{2})\)(?!\s+per month)", re.M)
repayment_date_reg = re.compile(r"by\s(\w+\s\d{1,2},\s\d{4})", re.M)
```

```python
## for loop to grab the targets in the confession docs
with open("confession-judgments-copy.txt", "r") as f:
    confession_text = f.read()

case_ids = case_num_reg.findall(confession_text)
date_received = date_received_reg.findall(confession_text)
total_amount = amt_reg.findall(confession_text)
repayment_full_date = repayment_date_reg.findall(confession_text)
```

```python
## make into df
df= pd.DataFrame(list(zip(case_ids, date_received, total_amount, repayment_full_date)))
df.columns = ["case_id", "date_received", "total amount", "repayment_full_date"]
df
```
